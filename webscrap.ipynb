```python
import pandas as pd
from bs4 import BeautifulSoup
from urllib.request import Request, urlopen
```

```python
alamat = "https://id.wikipedia.org/wiki/One_Piece"
req = Request(alamat)
```

```python
html = urlopen(req).read()
data = BeautifulSoup(html, 'html.parser')
table = data.find_all('table', {'class': 'wikitable'})[0]
```

```python
print(table)
```

```python
rows = table.findAll('tr')
```

```python
hasil = []
for row in rows:
   cols = row.find_all('td')
   cols = [ele.text.strip() for ele in cols]
   hasil.append(cols)
```

```python
df_hasil = pd.DataFrame(hasil)
df_hasil
```

```python
df_hasil = df_hasil.drop([0], axis=0) # menghapus baris pertama
df_hasil.columns = ['Judul', 'Tanggal Tayang'] # memberi judul kolom tabel
```

```python
df_hasil
```

```python
df_hasil.to_csv(r'D:\openapi\rawdata.csv', index=False)
```
